```python
def random_lst(lst):
    """Generates a random element from list"""
    return lst[random.randint(0,len(lst)-1)]
    
```

```python
def flight(player_def, player_lvl):
    """Determines if flight succeeds or fails"""
    dice_roll()
    slow_print(input("You have chosen to run\nTime to roll your die!\n"))
    slow_print("You rolled a:")
    die = random.randint(1,6)
    dice_result(die)
    if player_def + player_lvl + die >= 5:
        slow_print(f"\n"+flight_win[random.randint(0, len(flight_win)-1)])
        return 0 
    else:
        slow_print(f"\n"+flight_lose[random.randint(0, len(flight_lose)-1)])
        return -1 
```

```python
def loot():
    """ Gives player a choice of loot"""
    choice = ""
    while choice != "1" or choice != "2":
        if choice == "1" or choice == "2":
            break
        
        slow_print("\n\nYou defeat the monster and find: \n1: A sword (+1 strenght) \n2: A shield (+1 defence)\n")
        choice = input("")
    if choice == "1":
        return "1"
    elif choice == "2":
        return "2"
```

```python
def slow_print(str):
    """Prints text slower"""
    for letter in str:
        sys.stdout.write(letter)
        sys.stdout.flush()
        time.sleep(0.001)
```

```python
def slow_print_2(lst):
    """Prints text row one by one (list of list)"""
    for row in lst:
        for string in row:
            sys.stdout.write(string)
            sys.stdout.flush()
            time.sleep(0.1)
```
